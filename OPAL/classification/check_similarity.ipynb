```python
df_keyword_hier_similarity = fileLoader.load_data("files.df_keyword_hierarchical_similarity")
df_keyword_hier_similarity.index.names = [None] * df_keyword_hier_similarity.index.nlevels
np.fill_diagonal(df_keyword_hier_similarity.values, np.nan)
df_keyword_hier_similarity.head(3)
```

```python
# Welche column Namen sind in allen drei Dataframes enthalten?
common_columns = set(df_ai_similarity.columns) & \
                 set(df_keyword_cos_similarity.columns) & \
                 set(df_keyword_jaccard_similarity.columns) & \
                 set(df_keyword_hier_similarity.columns)
len(common_columns)
```

```python
# erzeuge für alle Dataframes die obere Dreiecksmatrix
def upper_triangle_matrix(df):
    mask = np.triu(np.ones(df.shape), k=1).astype(bool)
    upper_triangle_df = df.where(mask)
    return upper_triangle_df

df_ai_similarity_upper = upper_triangle_matrix(df_ai_similarity)
df_keyword_cos_similarity_upper = upper_triangle_matrix(df_keyword_cos_similarity)
df_keyword_jaccard_similarity_upper = upper_triangle_matrix(df_keyword_jaccard_similarity)
df_keyword_hier_similarity_upper = upper_triangle_matrix(df_keyword_hier_similarity)

df_keyword_jaccard_similarity_upper
```

```python
def sorted_pair(row):
    return tuple(sorted([row['file1'], row['file2']]))

for df in (df_ai, df_cos, df_jaccard, df_hier, df_ai):
    df['Pair'] = df.apply(sorted_pair, axis=1)
```
