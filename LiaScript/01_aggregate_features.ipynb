```python
def count_imports(row):
    if isinstance(row["import:"], float):
        return 0
    return len(row["import:"])

df_metadata["import_count"] = df_metadata.apply(count_imports, axis=1)
```

```python
def has_https_icon(row):
    if isinstance(row["icon:"], float):
        return ""
    if "http" in row["icon:"][0]:
        return row["icon:"][0]
    else:
        return ""

df_metadata["https_icon"] = df_metadata.apply(has_https_icon, axis=1)
```

```python
icon_list = df_metadata[df_metadata["https_icon"]!=""]["https_icon"].unique()
icon_list
```

```python
def has_https_logo(row):
    if isinstance(row["logo:"], float):
        return ""
    if "http" in row["logo:"][0]:
        return row["logo:"][0]
    else:
        return ""

df_metadata["https_logo"] = df_metadata.apply(has_https_icon, axis=1)
df_metadata[df_metadata["https_logo"]!=""]["https_logo"].unique()
```

```python
from matplotlib import pyplot as plt
import urllib.request 
import matplotlib.image as mpimg
from pathlib import Path
import cairosvg
import io
import numpy as np
from PIL import Image

def load_and_display_image(image_path):

    path = Path(image_path)
    file_extension = path.suffix.lower()
    
    try:
        if file_extension == '.svg':
            # Konvertiere SVG zu PNG im Speicher
            png_data = cairosvg.svg2png(url=str(path))
            image = Image.open(io.BytesIO(png_data))
            # Konvertiere zu RGB falls notwendig
            if image.mode != 'RGB':
                image = image.convert('RGB')
            # Konvertiere zu Numpy-Array
            image_array = np.array(image)
            
        elif file_extension == '.ico':
            # Öffne ICO mit PIL
            image = Image.open(path)
            # Nehme das erste Frame bei animierten ICOs
            if hasattr(image, 'n_frames'):
                image.seek(0)
            # Konvertiere zu RGB
            image = image.convert('RGB')
            # Konvertiere zu Numpy-Array
            image_array = np.array(image)
            
        else:  # .png, .jpg, .jpeg
            # Direkt mit Matplotlib laden
            image_array = mpimg.imread(path)
            # Konvertiere zu RGB falls RGBA
            if image_array.shape[-1] == 4:
                # Entferne Alpha-Kanal
                image_array = image_array[..., :3]

    except Exception as e:
        print(f"Error while loading {image_path}: {e}")
        return None

    return image_array
```

```python
fig, ax =plt.subplots(4,5, figsize=(10,10)) 

i=0
for icon in icon_list:
    file_extension = icon.split(".")[-1]
    print(icon)
    try:
        urllib.request.urlretrieve( icon, "image."+file_extension) 
    except:
        continue
    
    image = load_and_display_image("image."+file_extension)
    if image is None:
        continue

    row = i // 5
    col = i % 4
    ax[row, col].imshow(image)
    ax[row, col].axis("off")
    #ax[row, col].set_title(icon.split("/")[-1])
    i=i+1

print(i)
```
