```python
from pathlib import Path
import pandas as pd
```

```python
repo_data_folder = "/media/sz/Data/Connected_Lecturers/LiaScript/raw/"
repo_data_file = "LiaScript_repositories.p"
df_repos = pd.read_pickle(Path(repo_data_folder) / Path(repo_data_file))
```

```python
course_data_file = Path().resolve().parent / Path("data") / "LiaScript_courses"
df_courses = pd.read_pickle(Path(str(course_data_file)+".p"))
```

```python
df_repos['year']=df_repos['created_at'].dt.year
df_repos.groupby('year').size().plot(kind='bar', title='Number of liascript repositories created per year')
```

```python
df_courses['year']=df_courses['created_at'].dt.year
df_courses.groupby('year').size().plot(kind='bar', title='Number of documents created per year')
```
